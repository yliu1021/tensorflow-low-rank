```python
from glob import glob
import json
import pandas as pd
import numpy as np
from IPython.display import display_html
```

```python
results = [json.load(open(f)) for f in glob("set_rank_results_cifar10/*.json")]
results = pd.DataFrame(results)

results["best_loss"] = results["loss"].map(np.min)
results["best_loss_epoch"] = results["loss"].map(lambda x: np.argmin(x) + 1)
results["best_categorical_accuracy"] = results["categorical_accuracy"].map(np.max)
results["best_categorical_accuracy_epoch"] = results["categorical_accuracy"].map(lambda x: np.argmax(x) + 1)
results["best_val_loss"] = results["val_loss"].map(np.min)
results["best_val_loss_epoch"] = results["val_loss"].map(lambda x: np.argmin(x) + 1)
results["best_val_categorical_accuracy"] = results["val_categorical_accuracy"].map(np.max)
results["best_val_categorical_accuracy_epoch"] = results["val_categorical_accuracy"].map(lambda x: np.argmax(x) + 1)
results["initial_rank"].replace(-1, 256, inplace=True)
results["new_rank"].replace(-1, 256, inplace=True)
results.head()
```

```python
def visualize(column: str, rank_update_epoch: int):
    avg_results = pd.pivot_table(
        results.loc[
            results["rank_update_epoch"] == rank_update_epoch,
            ["initial_rank", "new_rank", column]
        ],
        index="initial_rank",
        columns="new_rank",
    )
    count_results = pd.pivot_table(
        results.loc[
            results["rank_update_epoch"] == rank_update_epoch,
            ["initial_rank", "new_rank", column]
        ],
        index="initial_rank",
        columns="new_rank",
        fill_value=0,
        aggfunc=len
    )
    display_html(
        avg_results
            .style
            .background_gradient(cmap='Blues', axis=None, vmin=0.77)
            .format("{:.3f}")
            .set_table_attributes("style='display:inline'")
            .set_caption("Avg")
            .set_properties(width="40px", height="40px")
            ._repr_html_() + \
        count_results
            .style
            .background_gradient(cmap='Blues', axis=None)
            .format("{:.0f}")
            .set_table_attributes("style='display:inline'")
            .set_caption("Count")
            .set_properties(width="40px", height="40px")
            ._repr_html_(),
        raw=True
    )
```

```python
for rank_update_epoch in np.sort(results["rank_update_epoch"].unique()):
    print(f"Rank updated on epoch: {rank_update_epoch}")
    visualize("best_val_categorical_accuracy", rank_update_epoch)
```
